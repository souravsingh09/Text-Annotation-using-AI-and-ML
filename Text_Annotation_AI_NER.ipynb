```python
# import all the necessary library for the operation.
import pandas as pd
import unidecode
import nltk
import re
import pandas as pd
import traceback
from nltk.corpus import stopwords
import json
import re
```

```python
# Opening JSON file
f = open('/Users/souravsingh/Desktop/response.json',)

# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
dataList = []
for i in range(len(data)):
    txt = re.sub('<[A-Za-z\/][^>]*>', '', data[i]['indications'])
    dataList.append(txt)
```

```python
unwanted = []
for i in range(len(dataList)):
    if len(dataList[i]) == 0:
        unwanted.append(i)
```

```python
print(len(dataList))
for ele in sorted(unwanted, reverse = True):
    del dataList[ele]

print(len(dataList))
```

```python
dataList[1]
```

```python
clean_Text = []
## read each element and clean the text
subst = ""
for text in dataList:
    # remove 1.1 headings and similar
    regex = r"(INDICATIONS AND USAGE )"
    text = re.sub(regex, subst, text, 0, re.MULTILINE)
    
    text = re.sub('INDICATIONS AND USAGE', ' ', text)
    
    regex = r"(\d\S\d+(.*?),,)"
    text = re.sub(regex, subst, text, 0, re.MULTILINE)
    
    ### renove ,,,,, from end of every element
    regex = r"(\S\,,,+)"
    text = re.sub(regex, subst, text, 0, re.MULTILINE)
    
    ### remove trademark 
    regex = r"(®)"
    text = re.sub(regex, subst, text, 0, re.MULTILINE)
    
    ### remove "" in continuation 
    regex = r"(\"\")"
    text = re.sub(regex, '" ', text, 0, re.MULTILINE)
    
    ## remove consecuitve commas
    regex = r"(\,\,)"
    text = re.sub(regex, ', ', text, 0, re.MULTILINE)
    
    ## 
    regex = r"(\*)"
    text = re.sub(regex, ', ', text, 0, re.MULTILINE)
    
    text = re.sub(r'\n', '', text)
    
    ## remove extra spaces
    text = re.sub(' +', ' ', text)

    word = unidecode.unidecode(text)
    
    clean_Text.append(word)
```

```python
disable_pipes =[pipe for pipe in nlp.pipe_names if pipe != 'ner']
```

```python
import random
from spacy.util import minibatch, compounding
from spacy import displacy
from pathlib import Path
```

```python
from spacy.training.example import Example
```

```python
lst= []
with nlp.disable_pipes(*disable_pipes):
        optimizer = nlp.resume_training()
        
        for itr in range(400):
            
            try:
  
                random.shuffle(train_data) # shuffle the training data before each iteration
                losses = {}
                batches = minibatch(train_data, size = compounding(1.0, 4.0, 1.001))

                for batch in batches:
                    
                    text, annotations = zip(*batch)
                    
                    example = []
                    # Update the model with iterating each text
                    for i in range(len(text)):
                        doc = nlp.make_doc(text[i])
                        example.append(Example.from_dict(doc, annotations[i]))

                    nlp.update(          
#                         text,
#                         annotations,
                        example,
                        drop = 0.5,  
                        sgd = optimizer,
                        losses = losses)

                    print('=======================================')
                    print('Interation = '+str(itr))
                    print('Losses = '+str(losses))
            
            except ValueError as ve:
                print("value error")
                print(ve)
                lst.append(ve)
#                 print(texts)
#                 break
         
```
